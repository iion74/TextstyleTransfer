```python
# ÌïÑÏöîÏãú ÏÑ§Ïπò
# pip install transformers scikit-learn matplotlib

import pandas as pd
import torch
from torch.utils.data import Dataset
from transformers import (
    T5Tokenizer,
    T5ForConditionalGeneration,
    Seq2SeqTrainer,
    Seq2SeqTrainingArguments
)
from sklearn.model_selection import train_test_split
from collections import Counter
import matplotlib.pyplot as plt

# ‚úÖ CUDA ÌôïÏù∏
print("‚úÖ CUDA ÏÇ¨Ïö© Í∞ÄÎä•:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("üñ•Ô∏è GPU:", torch.cuda.get_device_name(0))
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ‚úîÔ∏è 1. Îç∞Ïù¥ÌÑ∞ Î°úÎî©
df = pd.read_csv("C:/Users/AI-LJH/Desktop/cap/smilestyle_dataset_filtered.tsv", sep="\t")
df = df.dropna(thresh=3)

# ‚úîÔ∏è 2. ÏÉòÌîå ÏÉùÏÑ± (ÌîÑÎ°¨ÌîÑÌä∏ ÏòÅÏñ¥Ìôî + ÏµúÏÜå Í∏∏Ïù¥ ÌïÑÌÑ∞ÎßÅ)
def build_samples(df):
    samples = []
    for _, row in df.iterrows():
        row = row.dropna()
        for tgt_col in row.index:
            tgt_text = row[tgt_col]
            for src_col in row.index:
                if src_col == tgt_col:
                    continue
                src_text = row[src_col]
                if len(src_text.strip()) < 3 or len(tgt_text.strip()) < 3:
                    continue
                prompt = f"Translate to {tgt_col} style: {src_text}"
                samples.append((prompt, tgt_text))
    return samples

# ‚úîÔ∏è 3. ÌÜ†ÌÅ¨ÎÇòÏù¥Ï†Ä Î°úÎî© + pad_token Î≥¥Ïû•
tokenizer = T5Tokenizer.from_pretrained("C:/Users/AI-LJH/Desktop/cap/KETI-AIRke-t5-base")
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token  # ‚úÖ pad_token ÏÑ§Ï†ï
print("üîé pad_token_id =", tokenizer.pad_token_id)

# ‚úîÔ∏è 4. ÏÉòÌîå Î∂ÑÎ¶¨
all_samples = build_samples(df)
train_samples, val_samples = train_test_split(all_samples, test_size=0.1, random_state=42)

# ‚úîÔ∏è 5. Ïä§ÌÉÄÏùº Î∂ÑÌè¨ ÏãúÍ∞ÅÌôî
def plot_style_distribution(samples, title="Style Distribution"):
    counter = Counter(prompt.split()[2] for prompt, _ in samples)
    styles, counts = zip(*sorted(counter.items()))
    plt.figure(figsize=(12, 5))
    plt.bar(styles, counts)
    plt.xticks(rotation=45)
    plt.title(title)
    plt.ylabel("ÏÉòÌîå Ïàò")
    plt.tight_layout()
    plt.show()

plot_style_distribution(train_samples, "Train Style Distribution")
plot_style_distribution(val_samples, "Validation Style Distribution")

# ‚úîÔ∏è 6. Dataset ÌÅ¥ÎûòÏä§ Ï†ïÏùò
class PromptPairDataset(Dataset):
    def __init__(self, samples, tokenizer, max_length=64):
        self.samples = samples
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        src, tgt = self.samples[idx]
        model_inputs = self.tokenizer(
            src, max_length=self.max_length, truncation=True, padding="max_length", return_tensors="pt"
        )
        labels = self.tokenizer(
            tgt, max_length=self.max_length, truncation=True, padding="max_length", return_tensors="pt"
        )["input_ids"]
        labels[labels == tokenizer.pad_token_id] = -100  # ‚ùó pad ÎßàÏä§ÌÇπ
        model_inputs["labels"] = labels
        return {k: v.squeeze() for k, v in model_inputs.items()}

# ‚úîÔ∏è 7. Dataset ÏÉùÏÑ±
train_dataset = PromptPairDataset(train_samples, tokenizer)
val_dataset = PromptPairDataset(val_samples, tokenizer)

# ‚úîÔ∏è 8. Î™®Îç∏ Î°úÎî© + pad_token_id Î™ÖÏãú
model = T5ForConditionalGeneration.from_pretrained(
    "KETI-AIR/ke-t5-base",
    pad_token_id=tokenizer.pad_token_id
).to(device)

# ‚úîÔ∏è 9. ÌïôÏäµ ÏÑ§Ï†ï
training_args = Seq2SeqTrainingArguments(
    output_dir="C:/Users/AI-LJH/Desktop/cap/kot5_output_final",
    num_train_epochs=3,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=2,
    learning_rate=5e-4,
    logging_steps=50,
    save_steps=500,
    evaluation_strategy="steps",
    eval_steps=2000,
    predict_with_generate=True,
    save_total_limit=2,
    fp16=False,
    bf16=True,
    report_to="none"
)

# ‚úîÔ∏è 10. Trainer Íµ¨ÏÑ± Î∞è ÌïôÏäµ ÏãúÏûë
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset = PromptDataset(samples, tokenizer, device=device),
    eval_dataset=val_dataset,
    tokenizer=tokenizer
)

trainer.train(resume_from_checkpoint=False)

# ‚úîÔ∏è 11. Î™®Îç∏ Ï†ÄÏû•
trainer.save_model("C:/Users/AI-LJH/Desktop/cap/kot5_output_final/final_model")
tokenizer.save_pretrained("C:/Users/AI-LJH/Desktop/cap/kot5_output_final/final_model")
```

```python
# ‚úÖ ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏÑ§Ïπò ÌïÑÏöî Ïãú
# pip install transformers peft accelerate bitsandbytes

import os
import torch
import pandas as pd
from torch.utils.data import Dataset
from transformers import (
    T5Tokenizer,
    T5ForConditionalGeneration,
    Seq2SeqTrainer,
    Seq2SeqTrainingArguments,
    TrainerCallback
)
from peft import PromptTuningConfig, get_peft_model

# ‚úÖ CUDA ÌôïÏù∏
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("‚úÖ CUDA ÏÇ¨Ïö©:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("üñ•Ô∏è GPU:", torch.cuda.get_device_name(0))

# ‚úÖ Î™®Îç∏ Î°úÎìú (Í∏∞Î≥∏ Î™®Îç∏ + soft prompt)
base_model_id = "wisenut-research/KoT5-base"  

model = T5ForConditionalGeneration.from_pretrained(base_model_id)
tokenizer = T5Tokenizer.from_pretrained(base_model_id)
tokenizer.pad_token = tokenizer.eos_token

# ‚úÖ Soft Prompt ÏÑ§Ï†ï
peft_config = PromptTuningConfig(
    task_type="SEQ_2_SEQ_LM",
    prompt_tuning_init="TEXT",
    num_virtual_tokens=10,
    prompt_tuning_init_text="ÎãπÏã†ÏùÄ Îã§ÏñëÌïú ÎßêÌà¨Î•º ÌïôÏäµÌïú AIÏûÖÎãàÎã§.",
    tokenizer_name_or_path=base_model_id
)

model = get_peft_model(model, peft_config)
model.to(device)

# ‚úÖ Îç∞Ïù¥ÌÑ∞ Î°úÎî© Î∞è Ï†ÑÏ≤òÎ¶¨
df = pd.read_csv("C:/Users/AI-LJH/Desktop/cap/smilestyle_dataset_filtered.tsv", sep="\t")
df = df.dropna(thresh=3)

def build_samples(df):
    samples = []
    for _, row in df.iterrows():
        row = row.dropna()
        for tgt_col in row.index:
            tgt = row[tgt_col]
            for src_col in row.index:
                if src_col == tgt_col: continue
                src = row[src_col]
                prompt = f"Translate to {tgt_col} style: {src}"
                samples.append((prompt, tgt))
    return samples

samples = build_samples(df)

# ‚úÖ Dataset ÌÅ¥ÎûòÏä§ Ï†ïÏùò
class PromptDataset(Dataset):
    def __init__(self, samples, tokenizer, max_length=64, device="cpu"):
        self.samples = samples
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.device = device  # ‚úÖ Ï∂îÍ∞Ä

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        src, tgt = self.samples[idx]
        inputs = self.tokenizer(src, truncation=True, padding="max_length", max_length=self.max_length, return_tensors="pt")
        labels = self.tokenizer(tgt, truncation=True, padding="max_length", max_length=self.max_length, return_tensors="pt")["input_ids"]
        labels[labels == self.tokenizer.pad_token_id] = -100
        inputs["labels"] = labels

        # ‚úÖ Î™®Îì† ÌÖêÏÑúÎ•º GPUÎ°ú Î≥¥ÎÉÑ
        return {k: v.squeeze().to(self.device) for k, v in inputs.items()}



train_dataset = PromptDataset(samples, tokenizer)

# ‚úÖ 1000 stepÎßàÎã§ Ï∂úÎ†•ÌïòÎäî ÏΩúÎ∞±
class PrintEvery1000StepsCallback(TrainerCallback):
    def on_step_end(self, args, state, control, **kwargs):
        if state.global_step % 1000 == 0:
            logs = state.log_history[-1] if state.log_history else {}
            if "loss" in logs:
                print(f"\nüìå Step {state.global_step}/{state.max_steps} | Loss: {logs['loss']:.2f} | Grad Norm: {logs.get('grad_norm', 'N/A')}")

# ‚úÖ ÌïôÏäµ ÏÑ§Ï†ï
training_args = Seq2SeqTrainingArguments(
    output_dir="C:/Users/AI-LJH/Desktop/cap/kot5_softprompt_output",
    num_train_epochs=3,
    per_device_train_batch_size=8,
    learning_rate=5e-4,
    logging_steps=1000,  #
    save_steps=5000,
    eval_strategy="no",
    fp16=True,
    bf16=False, 
    report_to="none",
    disable_tqdm=True  # ü§´ tqdmÎèÑ ÎÅÑÍ∏∞
)

# ‚úÖ Trainer Íµ¨ÏÑ±
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset = PromptDataset(samples, tokenizer, device=device),
    tokenizer=tokenizer,
    callbacks=[PrintEvery1000StepsCallback()]  # ‚úÖ Ïª§Ïä§ÌÖÄ ÏΩúÎ∞± Ï∂îÍ∞Ä
)

# ‚úÖ ÌïôÏäµ ÏãúÏûë
trainer.train()

# ‚úÖ Î™®Îç∏ Ï†ÄÏû•
model.save_pretrained("C:/Users/AI-LJH/Desktop/cap/kot5_softprompt_output/final_model")
tokenizer.save_pretrained("C:/Users/AI-LJH/Desktop/cap/kot5_softprompt_output/final_model")
```
